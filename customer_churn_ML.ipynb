```python
# Copy of data
df=data.copy()
```

```python
# Droping necessary columns
df.drop(columns=['customerID'],inplace=True)
```

```python
df.head()
```

```python
# Checking dtypes
df.dtypes
```

```python
# Unique values in each column
cols=['gender','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract'
      ,'PaperlessBilling','PaymentMethod']
for col in df[cols]:
    print(col,df[col].unique())
```

### Data Preprocessing

```python
# Categorical data encoding 
```

```python
# One hot encoding 
```

```python
columns=['MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaymentMethod']
```

```python
df=pd.get_dummies(df,columns=columns,drop_first=True)
```

```python
df
```

```python
# Label Encoder
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
c=['gender','Partner','Dependents','PhoneService','PaperlessBilling']
for cols in df[c]:
    df[cols]=le.fit_transform(df[cols])
```

```python
df
```

```python
# Converting dtype object to float
```

```python
df.dtypes
```

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
# Handling missing values
```

```python
df.isna().sum()
```

```python
df.hist('TotalCharges') # Since the graph of Total Charges is skewed distribution, fill the missing values with median  
```

```python
df['TotalCharges']=df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
df.isna().any()
```

### Logistic Regression

```python
# Importing necessary libraries
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score
```

```python
# splitting train test data 
```

```python
x=df.drop(columns=['Churn'])
```

```python
y=df['Churn']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
# Training model
```

```python
lr=LogisticRegression()
```

```python
model=lr.fit(x_train,y_train)
```

```python
model
```

```python
# Prediction
pred=model.predict(x_test)
```

```python
accuracy_score(pred,y_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
print("Confusion matrix:")
print(confusion_matrix(y_test,pred))
```

```python
print("Classification Report:")
print(classification_report(y_test,pred))
```

```python
# Checking Overfitting/Underfitting
```

```python
train_score=model.score(x_train,y_train)
test_score=model.score(x_test,y_test)
```

```python
print("Train Score",train_score)
print("Test Score",test_score)
```

Model is performing well with no under or overfitting

### Random Forest Classifier

```python
# Copy of original data 
df2=data.copy()
```

```python
# Dropping necessary columns
df2.drop(columns=['customerID'],inplace=True)
```

```python
# Label Encoder
for col in df2.columns:
    df2[col]=le.fit_transform(df2[col])
```

```python
# Importing necessary libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
# Splitting train and test data
```

```python
x1=df2.drop(columns=['Churn'])
```

```python
y1=df2['Churn']
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,test_size=0.2,random_state=42)
```

```python
# Training model
```

```python
rf=RandomForestClassifier(max_depth=10,n_estimators=100)
```

```python
model2=rf.fit(x1_train,y1_train)
```

```python
model2
```

```python
# Prediction 
prediction=model2.predict(x1_test)
```

```python
accuracy_score(y1_test,prediction)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y1_test,prediction))
```

```python
print("Classification Report:")
print(classification_report(y1_test,prediction))
```

```python
# Checking Overfitting/Underfitting
```

```python
train_score2=model2.score(x1_train,y1_train)
test_score2=model2.score(x1_test,y1_test)
```

```python
print("Train Score",train_score2)
print("Test Score",test_score2)
```

Train 88% and Test 81% model is balanced
